```python
people = { "first" : ['nisheet','rohan','amit','abhi','priyanshu','shlok'],
            "last" : ['patel', 'soni', 'soni', 'patel','patel','bhavsar'],
          "email" : ["nisheet@gmail.com", "rohansoni@gmail.com","amitsoni@gmail.com","abhi@gmail.com",'priyanshu@gmail.com','sb22@gmail.com']
         }
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

### Using Conditionals
we can you >, <, ==, !=

```python
filt = (df['last'] == 'patel')
```

```python
# df[filt] or df[df['last'] == 'patel'] both are valid
df[filt]
```

```python
df.loc[filt]
```

```python
# It gives email of only with patel == last
df.loc[filt, 'email']
```

```python
# & = and
# | = or
df[(df['last'] == 'patel') & (df['first'] == 'nisheet')]
```
